## reading csvs

````python
import pandas as pd


def read_csv(path):
    df = pd.read_csv(path)

    # cast typings
    for col in df.columns[1:]:
        df[col] = df[col].astype(float)
    df.Date = pd.to_datetime(df.Date)
    
    return df

df_prod_test = read_csv('Test.csv')
df = read_csv('Data.csv')
original_columns = df.columns # this will come in handy
# split train and test (no time shuffling)
# before actual testing we'll come up with model using these datasets
# then retrain model on all available data before actual run on prod test data
def split_df(df, train_fraction=0.75):
    dates = pd.Series(df.Date.unique()).sort_values()
    df_train = df[df.Date < dates[int(len(dates) * train_fraction)]]
    df_test = df[df.Date >= dates[int(len(dates) * train_fraction)]]
    return df_train, df_test

df_train, df_test = split_df(df)

df_train
````

## So in each row all price columns are fairly tightly distributed with few missing values in Bid/Ask columns and with outliers

## Its very important to note that TargetPrice column is almost always very close in value to median of price columns

## At this point its obvious to me that these rows are observations for _different_ assets that all have different long-term mean price levels. Therefore for the model to generalize prices should be normalized in a way that brings down this price level difference between assets
#### In general its more to it than this obviously. eg maybe its worth throwing in volatility into normalization but its not possible with this dataset without instruments labeled and unknown nature of features


# Considered the data damage situation I'm going to normalize price columns by dividing them by FeaturePrice1 as it has the least amount of data loss. And in that cases where FeaturePrice1 is 0 or outlier the price_norm is equal to median(price columns in row)

````python
def get_price_norm(row):
    if row.FeaturePrice1 < row.price_median * 2 \
            or row.FeaturePrice1 > row.price_median / 2:
        # if FeaturePrice1 is outlier or 0
        return row.price_median
    # else return FeaturePrice1
    return row.FeaturePrice1

def add_price_norm(df):
    price_columns = [col for col in df.columns
                     if 'Price' in col]
    df['price_median'] = df[price_columns].median(axis=1)
    df['price_norm'] = df.apply(get_price_norm, axis=1)
    
add_price_norm(df)
add_price_norm(df_prod_test)
df
````

## For the same reasoning Im going to multiply `Bid_Shares` and `Ask_Shares` by price
# And following histograms demonstrate why I think its worth doing

````python
import matplotlib.pyplot as plt

def plot_hist(series, limit_quantile=0.9, bins=200):
    q_limited = series[series < series.quantile(limit_quantile)]
    q_limited.hist(bins=bins)

print('Shares')
plot_hist(df.Bid_Shares)

plt.show()

print('Dollars')
plot_hist(df.Bid_Shares * df.Bid_Price)
````

````python
def add_bid_ask_quote(df):
    df['Bid_quote'] = df.Bid_Shares * df.apply(
        lambda row: row.Bid_Price if row.Bid_Price else row.price_norm,
        axis=1)
    df['Ask_quote'] = df.Ask_Shares * df.apply(
        lambda row: row.Ask_Price if row.Ask_Price else row.price_norm,
        axis=1)

add_bid_ask_quote(df)
add_bid_ask_quote(df_prod_test)
````

## Im willing to throw out these rows where TargetPrice is far away from median of price columns. Either they are too damaged or actual TargetPrice is damaged. Anyways there are really not much of them and they are hurting learning and model evaluation process
### I'm going to assume that in production scenario there are good ways to handle these. eg not take action when you get too big prediction and wait for next observation

````python
src_df = df.copy()
# normalizing price_columns
for col in price_columns:
    src_df[col] /= src_df['price_norm']

src_df[(src_df.TargetPrice > 2) | (src_df.TargetPrice < 0.5)]
````

## Prepare datasets for ML

````python
src_df = src_df[(src_df.TargetPrice <= 2) & (src_df.TargetPrice >= 0.5)]

df_train, df_test = split_df(src_df)
# replace bid_ask_shares with bid_ask_usd
x_cols = list(original_columns[1:-1])
x_cols.remove('Bid_Shares')
x_cols.remove('Ask_Shares')
x_cols.extend(['Bid_quote', 'Ask_quote'])

x_df_train = df_train[x_cols]
y_df_train = df_train['TargetPrice']

x_df_test = df_test[x_cols]
y_df_test = df_test['TargetPrice']

x_df_train
````

## Lets also normalize `FeatureShares2` with standard scaler and `FeatureShares1`, `Bid_quote`, `Ask_quote` with MinMax

````python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

def scale_std_and_minmax(df_train, df_test=None,
                         std_scaler=None, minmax_scaler=None):
    std_cols = ['FeatureShares2']
    if std_scaler is None:
        std_scaler = StandardScaler().fit(df_train[std_cols])
    df_train.loc[:, std_cols] = std_scaler.transform(df_train[std_cols])

    minmax_cols = ['FeatureShares1', 'Bid_quote', 'Ask_quote']
    if minmax_scaler is None:
        minmax_scaler = MinMaxScaler().fit(df_train[minmax_cols])
    df_train.loc[:, minmax_cols] = minmax_scaler.transform(df_train[minmax_cols])
    df_train.loc[:, std_cols] = std_scaler.transform(df_train[std_cols])

    if df_test is not None:
        df_test.loc[:, minmax_cols] = minmax_scaler.transform(df_test[minmax_cols])
        df_test.loc[:, std_cols] = std_scaler.transform(df_test[std_cols])

    return std_scaler, minmax_scaler

std_scaler, minmax_scaler = scale_std_and_minmax(x_df_train, x_df_test)
````

## Now throw this in `GradientBoostingRegressor`
##### I selected this model after trying others and it works best. For this particular problem I havent got much preferences in model selection. Obviously there is not enough data for deep learning. GradientBoosting should work well with one-hot data and although there is none now but it can be added to this dataset AND I think the model actually handles that zeros nicely. In dev environment I usually use auto ml framework like `tpot` to bruteforce through models and see if its possible to get an edge


````python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error

import numpy as np
np.random.seed(42) # Reproducibility

x_train, y_train = x_df_train.values, y_df_train.values
x_test, y_test = x_df_test.values, y_df_test.values

reg = GradientBoostingRegressor()
reg.fit(x_train, y_train)

mean_absolute_error(y_test, reg.predict(x_test))
````

## Looks nice but we care about abs percentage error of taget prices denormalized back. Lets check it out!

````python
dft = df_test.copy()

dft['TargetPriceRaw'] = dft.TargetPrice * dft.price_norm
dft['pred'] = reg.predict(x_test)
dft['pred_raw'] = dft.pred * dft.price_norm
dft['error'] = (
    (dft.TargetPriceRaw - dft.pred_raw).abs()
    / dft.TargetPriceRaw
)
print(f'MAPE between target price and prediction is {round(dft.error.mean() * 100, 3)}%')
dft[['error']].describe()
````

````python
dft.sort_values('error', ascending=False).head(20)
````

## Inspection shows there is still the "data damage" in play

# Lets retrain the model on all data and predict the Test.csv

````python
df_prod_train = src_df.copy()
x_df_train = df_prod_train[x_cols]
y_df_train = df_prod_train['TargetPrice']

std_scaler, minmax_scaler = scale_std_and_minmax(x_df_train)

np.random.seed(42) # Reproducibility
x_train, y_train = x_df_train.values, y_df_train.values

reg = GradientBoostingRegressor()
reg.fit(x_train, y_train)


df_prod_test_norm = df_prod_test.copy()
# normalizing price_columns
for col in price_columns:
    if col in df_prod_test_norm.columns:
        df_prod_test_norm[col] /= df_prod_test_norm['price_norm']

x_df_test = df_prod_test_norm[x_cols]

scale_std_and_minmax(x_df_test, None,
                     minmax_scaler=minmax_scaler,
                     std_scaler=std_scaler)

df_prod_test['_pred_raw'] = pd.Series(reg.predict(x_df_test)) \
                              .clip(lower=0.5, upper=2)
df_prod_test['TargetPrice'] = df_prod_test['_pred_raw'] * df_prod_test.price_norm

df_prod_test
````
